я решила не добавлять ничего более конекретного, чтобы было проще приводить к единому стандарту и сравнивать разметки парсеров

ответы парсеров сразу же сводятся к единому формату с помощью ргеулярных выражений и только в таком виде попадают в переменные _answer

```python
def the_smart_guy(tagger, gold=gold): #функция, считающая accuracy
    gold_POSs = []
    for i in gold:
        gold_POSs.append(i[1])
        
    POSs = []
    for i in tagger:
        POS = list(i.values())[0]
        POSs.append(POS)
        
    mistakes = 0
    i = 0
    for one in POSs:
        if not gold_POSs[i] in one:
            mistakes += 1
        i += 1
    accuracy = (len(POSs)-mistakes)/len(POSs)*100
    return accuracy
```

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
pymorphy_answer = []

for word in words:
    mytags = morph.parse(word)
    myPOSs=[]
    for mytag in mytags:
        myPOS = mytag.tag.POS
        myPOS = re.sub(r'INFN', 'VERB', myPOS)
        myPOS = re.sub(r'ADJ.', 'ADJ', myPOS)
        myPOS = re.sub(r'PRT.', 'PRT', myPOS)
        myPOS = re.sub(r'ADVB', 'ADV', myPOS)
        myPOS = re.sub(r'NPRO', 'PRO', myPOS)
        if not myPOS in myPOSs:
            myPOSs.append(myPOS)
    
    word_dic = dict.fromkeys([word], myPOSs)
    pymorphy_answer.append(word_dic)
    
print(pymorphy_answer)
print(len(pymorphy_answer))
```

```python
from natasha import (Segmenter, NewsEmbedding, NewsMorphTagger, NewsSyntaxParser, Doc)

segmenter = Segmenter()
emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
doc = Doc(text)

doc.segment(segmenter)
doc.tag_morph(morph_tagger)

natasha_answer = []

for item in doc.tokens:
    word = re.findall('[а-я-]+', str(item))
    myPOS = re.search('[A-Z]{2,}', str(item))[0]
    myPOS = re.sub(r'ADP', 'PREP', myPOS)
    myPOS = re.sub(r'PART', 'PRCL', myPOS)
    myPOS = re.sub(r'CCONJ', 'CONJ', myPOS)
    myPOS = re.sub(r'NUM', 'NUMR', myPOS)
    myPOS = re.sub(r'PRON', 'PRO', myPOS)
    myPOS = re.sub(r'DET', 'PRO', myPOS)
    word_dic = dict.fromkeys(word, myPOS)
    natasha_answer.append(word_dic)

print(natasha_answer)
print(len(natasha_answer))
```

```python
from pymystem3 import Mystem
m = Mystem()
mystem_answer = []

for word in words:
    mytag = m.analyze(word)
    myPOS = mytag[0]['analysis'][0]['gr']
    myPOS = re.search('[A-Z]+', myPOS)[0]
    
    myPOS = re.sub('PART', 'PRCL', myPOS)
    myPOS = re.sub('NOUNPRO', 'PRO', myPOS)
    myPOS = re.sub('SPRO', 'PRO', myPOS)
    myPOS = re.sub('ADVPRO', 'PRO', myPOS)
    if myPOS == 'A':
        myPOS = re.sub('A', 'ADJ', myPOS)
    if myPOS == 'S':
        myPOS = re.sub('S', 'NOUN', myPOS)
    if myPOS == 'PR':
        myPOS = re.sub('PR', 'PREP', myPOS)
    if myPOS == 'V':    
        myPOS = re.sub('V', 'VERB', myPOS)
    
    word_dic = dict.fromkeys([word], myPOS)
    mystem_answer.append(word_dic)
    
print(mystem_answer)
print(len(mystem_answer))
```

Про улучшение программы из предыдущей домашки:

1. можно (и даже очень нужно) смотреть на сочетания "не ADJ / ADV / VERB" (например, о фильмах: "не интересный" / "не смешно" / "не захватывает" или, наоборот, "не плохой" / "не нудно" / "не разочаровал"), потому что в отрыве от "не" такие словосочетания значат противоположное, что очевидно, всё рушит.

2. ADJ + NOUN / NOUN + ADJ (или даже NOUN + "очень" + ADJ): так можно сразу вылавливать что-то вроде "интересный сюжет", "хороший каст", "режиссёр талантливый", "музыка очень крутая".

3. сочетание "мне" + VERB встречается, наверное, в отзывах на любой продукт и, что важно, коротко и ясно описывает тональность собственно отзыва (автору либо нравится, и отзыв положительный, либо не нравится, и отзыв отрицательный)
